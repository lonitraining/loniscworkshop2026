# **Chapter 1: Why are we here?**

We are living in a time where technology, coding, and data analysis are changing very quickly. Today, AI tools can help us write code in languages like R and Python. So, it is fair to ask: “Why are we here? Do we still need to learn R?”

## **AI is fast, but can be wrong**
The answer is yes. AI can write R code quickly, explain errors, and suggest solutions. This can save time and make learning easier. However, fast code does not always mean correct code. Coding is only one part of data analysis.

![picture1](https://raw.githubusercontent.com/kasettisiva/R_training_summer_workshop_2026/main/images/coding_with_ai_cartoon_1.png)

AI-generated code may contain errors, use the wrong method, or miss important details in your data. It may not fully understand your research question, outliers, assumptions, or domain context. That is why you are responsible for checking whether the code works correctly.

![picture2](https://raw.githubusercontent.com/kasettisiva/R_training_summer_workshop_2026/main/images/coding_with_ai_cartoon.png)

Your expertise is vital: know what to ask, rigorously verify AI's output, and accurately interpret results. This human oversight is key to effective collaboration with AI and prevents misleading outcomes.

The best way to use AI is as a helper, not as a replacement for your understanding. A good workflow is:

## **Few Ways AI Can Help**

That said, AI is genuinely useful — when you know how to use it. Here are five ways it can help you right now.

### **a) Explain Code**

Prompt: *"Explain this R code in simple words: `mean(mtcars$mpg)`"*

```r
mean(mtcars$mpg)   # average miles per gallon from the built-in mtcars dataset
```

Use AI explanations to learn what each function does, not just to get a number.

### **b) Write Starter Code**

Prompt: *"Write R code to find the average Sepal.Length per Species in iris using dplyr."*

```r
library(dplyr)

iris %>%
  group_by(Species) %>%
  summarize(avg_sepal_length = mean(Sepal.Length))
```

Check before you trust. Are column names correct? Does the result make sense? Can you explain it?

### **c) Debug Errors**

Give AI context + code + error — you get a fix and an explanation.

```
I am working in R. My goal is to calculate average Sepal.Length in iris.

Code:  mean(iris$SepalLength)
Error: [paste the error message]

Explain the issue simply and give corrected code.
```

```r
# This breaks — wrong column name
# mean(iris$SepalLength)

# Correct
mean(iris$Sepal.Length)
```

### **d) Explain Statistical Output**

AI can decode statistical output — but you must decide which test is appropriate.

```r
t.test(mpg ~ am, data = mtcars)
```

Prompt: *"Explain this t-test output: what do the p-value, confidence interval, and group means mean?"*

AI explains the numbers. You decide if the test fits the research question.

### **e) Improve and Refactor Code**

Prompt: *"Make this R code more readable and give the output column a better name."*

```r
# Before: dense, unnamed output column
# iris %>% group_by(Species) %>% summarize(mean(Petal.Length))

# After: readable, meaningful names
iris %>%
  group_by(Species) %>%
  summarize(
    mean_petal_length = mean(Petal.Length)
  )
```

## **How to write good prompts for R code?**

❌ **Vague prompt**
> "Write R code to analyze my data"

✅ **Specific prompt**
```
I am working in R.

My goal is:     [what you want to do]
My code is:     [paste code]
The problem is: [error message or unexpected output]

Explain the issue simply and give corrected R code.
```

More context → better answer. Vague prompts get vague code. The better your prompt describes your data structure and goal, the more accurate the generated code will be.

## **Common AI code review checklist**

Before using AI-generated R code, verify:

- [ ] **Does it run?** — Paste and execute; don't assume it works
- [ ] **Are the variable names right?** — AI sometimes invents column names
- [ ] **Is `na.rm = TRUE` needed?** — AI often forgets missing value handling
- [ ] **Is the package loaded?** — AI often uses functions without loading the library
- [ ] **Is the method appropriate?** — AI picks common methods, not necessarily the *right* one
- [ ] **Does the result match expectations?** — Sanity-check with `summary()` and a plot
- [ ] **Are you sharing private data?** — Never paste real patient/proprietary data into AI tools

## **What AI is good at?**

| What AI is good at | What YOU still need |
|---|---|
| Writing boilerplate code | Understanding what the code does |
| Suggesting packages | Knowing whether the approach is valid |
| Fixing syntax errors | Catching logic errors and wrong assumptions |
| Generating plots | Interpreting what the plot means |
| Explaining concepts | Deciding which method fits your data |

> AI is a *coding assistant*, not actually doing analysis or taking decisions.

## **AI is just a tool**
AI cannot replace your domain knowledge, your understanding of the data, or your judgment about what the results mean. Use it as a capable assistant — not as an authority.

> **The bottom line:** Learning R in the AI era is not about memorizing syntax. It is about understanding your data, your question, and your results well enough to catch what AI gets wrong.
