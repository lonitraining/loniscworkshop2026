**Dataset:** The Palmer Penguins dataset contains body measurements of **344 penguins**
collected between 2007 and 2009 at three islands in the Palmer Archipelago, Antarctica.
There are **three species**: *Adelie*, *Chinstrap*, and *Gentoo*.

It is a friendly, real-world dataset — a modern alternative to the famous `iris` dataset.

| Column | What it means |
|--------|---------------|
| `species` | Penguin species (Adelie / Chinstrap / Gentoo) |
| `island` | Island where observed (Biscoe / Dream / Torgersen) |
| `bill_length_mm` | Length of the bill (beak) in mm |
| `bill_depth_mm` | Depth (thickness) of the bill in mm |
| `flipper_length_mm` | Flipper length in mm |
| `body_mass_g` | Body mass in grams |
| `sex` | Sex (male / female) |
| `year` | Year of observation (2007, 2008, or 2009) |
