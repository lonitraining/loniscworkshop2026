# **1. What is R**

* R is a programming language for statistical computing
  * Importing, storing, exporting and manipulating data
  * Conducting statistical analyses
  * Displaying the results by tables, graphs, etc.
* R is also a software environment for the development and implementation of new algorithms.
  * Many graphical user interface to R both free and commercial (e.g. Rstudio and Microsoft R Open).
* R is being used by many disciplines
  * [A collection of repositories of R codes categorized by discipline](https://github.com/lsuhpchelp/r_collection)

## **1.1 Features of R**

* Designed for statistical analysis, with rich data analysis functionalities and sophisticated graphical capabilities
* Available on most platform/OS
* Active development and (very) active community
  * [CRAN](https://cran.r-project.org/): The Comprehensive R Archive Network
  * Source code and binaries, user contributed packages and documentation
  * More than 17,000 packages available on CRAN (as of May 2021)
    * Compared to 6,000 four years ago
* Free to use

## **1.2 R's Unique Strengths**

- **Designed for statistical computing and graphics**  
  R is officially described as a language and environment for statistical computing and graphics. This makes it a natural choice for data analysis, statistical tests, modeling, and visualization. Learn more: https://quarto.org/docs/get-started/

- **Strong built-in statistical tools**  
  R includes many standard statistical functions, such as `t.test()`, `lm()`, and `aov()`, which are commonly used for hypothesis testing, regression, and analysis of variance.

- **Powerful data visualization with ggplot2**  
  `ggplot2` is based on the Grammar of Graphics. You provide the data, map variables to visual features, choose the type of plot, and ggplot2 handles the details. Learn more: https://quarto.org/docs/get-started/

- **Widely used in bioinformatics and genomics**  
  Bioconductor provides open-source software for precise and repeatable analysis of biological data, making R especially useful for genomics, gene expression, and other life science workflows. Learn more: https://quarto.org/docs/get-started/

- **Reproducible reports with Quarto or RMarkdown**  
  Quarto supports documents that combine code, results, plots, and written explanation. Reports can be regenerated when the data or analysis changes. :contentReference[oaicite:3]{index=3}

- **Works well in HPC workflows**  
  R scripts can be submitted as regular SLURM jobs. This makes it easy to run statistical analysis, simulations, job arrays, and parallel workloads on HPC systems.

## **2.1 Basic Syntax**


### **a) Assignment Operator**

For new users, the biggest difference from other languages is perhaps the assignment operator: R uses `<-` instead of `=`.

````r
#@markdown **Examples**
#R convention: use <- for assignment

gene_count <- 24000   # Numeric
sample_id  <- "SRR123456"  # Character
is_paired  <- TRUE    # Logical

# Print values
gene_count
sample_id
is_paired
````

### **b) Legal R Names**

Names for R objects can be any combination of letters, numbers, underscores (_) and periods (.), but must not start with a number or an underscore.

````r
#@markdown **Examples**

# This is legal.
sample_id <- "S001"
sample_id

# This is legal too.
sample.ID.1 <- "S002"
sample.ID.1

# This is NOT legal because it starts with a number.
# 2026_sample <- "Blood sample"

# This is NOT legal because it starts with underscore.
# _2026_sample <- "Blood sample"

# This is NOT recommended because spaces are not allowed in normal object names.
# sample id <- "S001"
````

## **2.3 Getting help**
Getting help is straightforward in R.

For information about specific functions or objects, use `?<name of funciton>` or `help(<name of function>)`.

```r
?mean           # help page for a function
help(mean)      # another way to get help
??normalization  # search for keyword across all installed packages
help.search("normalization") # another way to search for keywords
example(mean)   # show examples of how to use a function
```

> 💡 **Tip:** If the help page is confusing, paste the function signature into Claude or ChatGPT and ask "explain this in plain English with an example."
