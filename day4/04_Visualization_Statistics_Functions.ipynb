# **5 · Visualization with ggplot2**


### **5.1 Base R Graphics: Quick and Direct**

R offers built-in plotting functions for direct and immediate creation of basic plots. They are great for quick data exploration, building plots incrementally by adding elements to the canvas.



**Formula for `plot()` function**

The `plot()` function in Base R is a generic function for plotting R objects. Its behavior depends on the class of the first argument. However, for basic x-y plotting (scatter plots, line plots), the general syntax is:

```
plot(x, y, ...)
```

Where:
* `x`: The values on the horizontal axis.
* `y`: The values on the vertical axis.
* `...`: Optional arguments to control plot appearance, such as:
    * `main`: Main title of the plot.
    * `xlab`: Label for the x-axis.
    * `ylab`: Label for the y-axis.
    * `type`: Type of plot (e.g., `'p'` for points, `'l'` for lines, `'b'` for both).
    * `col`: Color of plotting symbols or lines.
    * `pch`: Plotting character (symbol) for points.
    * `cex`: Character expansion factor (size of plotting symbols).
    * `lwd`: Line width.
    * `xlim`: x-axis limits (e.g., `c(min, max)`).
    * `ylim`: y-axis limits (e.g., `c(min, max)`).


### **5.2 Introducing `ggplot2`: The Grammar of Graphics**

`ggplot2` is a powerful and preferred tool for complex, publication-quality plots, based on the "grammar of graphics." This systematic approach builds plots in layers:

*   **Data:** The dataset.
*   **Aesthetics (`aes`):** Mapping data variables to visual properties (e.g., `x`, `y`, `color`).
*   **Geometries (`geom_*`):** Visual representation (e.g., points, lines, bars).
*   **Facets (`facet_*`):** Sub-plots based on categories.
*   **Statistics (`stat_*`):** Data transformations.
*   **Scales:** Control aesthetic mappings.
*   **Themes (`theme_*`):** Non-data elements (fonts, background).



The basic structure for creating a plot is:

```R
ggplot(data = <DATA>, aes(x = <X_VARIABLE>, y = <Y_VARIABLE>, ...)) +
  <GEOM_FUNCTION>(mapping = aes(...), stat = ..., position = ...) +
  <COORDINATE_FUNCTION>() +
  <FACET_FUNCTION>() +
  <STAT_FUNCTION>() +
  <SCALE_FUNCTION>() +
  <THEME_FUNCTION>() +
  <LABELS_FUNCTION>()
```

**Key Components Explained:**

1.  **`ggplot(data = ..., aes(...))`**: This is the initial call that creates a plot object. You specify:
    *   `data`: The dataset (a data frame) to be used for the plot.
    *   `aes(...)`: Aesthetic mappings. This links variables from your data to visual properties of the plot (e.g., `x` for x-axis, `y` for y-axis, `color`, `fill`, `size`, `alpha`, `shape`).

2.  **`<GEOM_FUNCTION>()`**: Geometric objects (geoms) represent the visual elements of the plot. You add one or more geoms using the `+` operator. Common examples include:
    *   `geom_point()`: For scatter plots.
    *   `geom_line()`: For line plots.
    *   `geom_bar()`: For bar charts.
    *   `geom_histogram()`: For histograms.
    *   `geom_boxplot()`: For box plots.
    *   `geom_smooth()`: For adding smoothed conditional means (e.g., regression lines).
    *   Most geoms accept additional arguments for their appearance (e.g., `color`, `fill`, `size`, `alpha`).

3.  **`<LABELS_FUNCTION>()`**: Functions to add titles, subtitles, and axis labels.
    *   `labs(title = "...", subtitle = "...", x = "...", y = "...")`: Customize labels.

4.  **`<THEME_FUNCTION>()`**: Controls the non-data visual elements of the plot, such as fonts, backgrounds, grid lines, and legend appearance.
    *   `theme_minimal()`: A minimalist theme.
    *   `theme_bw()`: A black and white theme.
    *   `theme()`: For fine-grained control over individual theme elements.

5.  **`<FACET_FUNCTION>()`**: Splits the plot into sub-plots based on the levels of one or more categorical variables.
    *   `facet_wrap(~ <VARIABLE>)`: Creates a series of plots, wrapping them into rows/columns.
    *   `facet_grid(<ROW_VARIABLE> ~ <COL_VARIABLE>)`: Arranges plots in a grid defined by two variables.

6.  **`<SCALE_FUNCTION>()`**: Controls how aesthetic mappings are translated into visual properties (e.g., specific colors for different categories, limits of axes).
    *   `scale_color_manual(values = c(...))`: Assign specific colors.
    *   `scale_x_continuous(limits = c(...))`: Set x-axis limits.

7.  **`<STAT_FUNCTION>()`**: Statistical transformations that summarize the data before plotting (e.g., calculating counts, means).
    *   Often, geoms have a default `stat` (e.g., `geom_bar()` uses `stat_count()`). You can override this if needed.

### **5.3 Why `ggplot2` is Widely Used:**

1.  **Modularity & Consistency:** Layered grammar makes plots logical, easy to modify, and visually consistent.
2.  **Aesthetics & Professionalism:** Produces publication-ready plots with less effort.
3.  **Flexibility & Customization:** Highly flexible for adding transformations, custom themes, and arranging plots.
4.  **Data-centric:** Promotes good visualization practices by focusing on mapping data to visuals.

**Key Differences:**

*   **Base R:** Direct, imperative plotting; good for quick, basic plots. Can be less intuitive for complex, multi-layered graphics.
*   **`ggplot2`:** Declarative, grammar-based plotting; ideal for sophisticated, customizable, and aesthetically superior visualizations. It has a steeper initial learning curve but provides greater control and consistency in the long run.

Here are the few examples of plotting with `ggplot2`.

````r
# Scatter plot: weight vs fuel efficiency, colored by cylinders
ggplot(mtcars, aes(x = wt, y = mpg, color = factor(cyl))) +
  geom_point(size = 3, alpha = 0.8) +
  geom_smooth(method = "lm", se = FALSE) +  # regression line per group
  labs(
    title = "Fuel Efficiency vs Weight",
    subtitle = "By number of cylinders",
    x = "Weight (1000 lbs)",
    y = "Miles per Gallon",
    color = "Cylinders"
  ) +
  theme_minimal()
````

````r
# Box plot: distribution of mpg by cylinder count
ggplot(mtcars, aes(x = factor(cyl), y = mpg, fill = factor(cyl))) +
  geom_boxplot(alpha = 0.7, outlier.shape = 21) +
  labs(title = "MPG Distribution by Cylinder Count",
       x = "Cylinders", y = "Miles per Gallon") +
  theme_bw() +
  theme(legend.position = "none")
````

## **6.2 Correlation**

Correlation tells us how strongly two numeric variables are related.

For example, you can ask the question:
> Do heavier cars have lower fuel efficiency?

````r
# How strongly are weight and fuel efficiency correlated?
cor(mtcars$wt, mtcars$mpg)
# r = -0.87 → strong negative correlation (heavier cars get worse mileage)
````

## **6.3 Linear Regression**

Linear regression is used when we want to predict one numeric variable using another variable.

For example:

> Can we predict a car’s MPG using its weight?

````r
# Simple linear regression: predict mpg from weight
model_lm <- lm(mpg ~ wt, data = mtcars)
summary(model_lm)
````

Multiple regression uses more than one predictor.

For example:

> Can we predict MPG using weight, horsepower, and number of cylinders?

````r
# Multiple regression: add horsepower and cylinders
model_multi <- lm(mpg ~ wt + hp + cyl, data = mtcars)
summary(model_multi)

# Key things to read in summary():
# - Estimate: slope for each predictor
# - Pr(>|t|): p-value (< 0.05 is conventionally "significant")
# - R-squared: % of variance explained (0.84 = 84% here)
# - Residual standard error: typical prediction error
````

## **6.4 t-test: comparing two groups**

A t-test compares the average value between two groups.

For example:

> Do 4-cylinder cars and 8-cylinder cars have different average MPG?

````r
# Do 4-cylinder and 8-cylinder cars differ significantly in MPG?
cars_4cyl <- mtcars[mtcars$cyl ==4, "mpg"]
cars_8cyl <- mtcars[mtcars$cyl ==8, "mpg"]

t.test(cars_4cyl, cars_8cyl)
# p < 0.05 → yes, the difference is statistically significant
````

## **7.2 The ```plyr``` package**

Suppose that, with the **mtcars** data frame, we want to know the average mileage-per-gallon for cars with 4, 6 and 8 cylinders. How do we do that?

We will need to
* **split** the data into subsets according to the value of the "cyl" column (one for cyl==4, one for cyl==6 and one for cyl==8)
* **apply** the mean function to the "mpg" column
* **combine** the results from each subset

````r
# Subset where cyl==4
mean(mtcars[mtcars$cyl==4,"mpg"])
# Subset where cyl==6
mean(mtcars[mtcars$cyl==6,"mpg"])
# Subset where cyl==8
mean(mtcars[mtcars$cyl==8,"mpg"])
````

The "split-apply-combine" pattern is very common in data analysis, where you solve a complex problem by breaking it down into small pieces, doing something to each piece and then combining the results back together again.

The `plyr` packages provide a group of functions that implement this split-apply-combine pattern.

For example, the `ddply()` function takes a data frame, split it accorindg to the condition you supply, apply a function to each piece, then combine the result into a new data frame.

````r
library(plyr)
ddply(mtcars,"cyl", summarize, AverageMPG=mean(mpg))
````

The "split" step can be done according to more than one variable.

The command below will tell us the average MPG for each unique combination of "gear" and "cyl":

````r
# Find the average mpg for each unique combination of "gear" and "cyl".
ddply(mtcars,c("gear","cyl"), summarize, AverageMPG=mean(mpg))
````

# **9 . Parallel Processing**

Modern computers are equipped with more than one CPU core and are capable of processing workloads in parallel, but base R is single‐threaded, i.e. not parallel.

In other words, regardless how many cores are available, R can only
use one of them.

There are two options to run R in parallel: **implicit** and **explicit**.


## **9.1 Implicit parallel processing**

Some functions in R can call parallel numerical libraries.

For instance, on the LONI QB-2 and QB-3 clusters most linear algebraic and related functions (e.g. linear regression, matrix decomposition, computing inverse and determinant of a matrix) leverage the multi‐threaded Intel MKL library.

In this case, no extra coding is needed to take advange of the multiple CPU cores - those functions will automatically use multiple cores when being called.

## **9.2 Explicit parallel processing**

If the implicit option is not available for what you'd like to do, some codes need to be written.

Here is an example of using the `%dopar%` directive in the `doParallel` package.

The workload is to generate 100 random samples, each with
1,000,000 observations from a standard normal distribution, then take a summary for each sample.

````r
iters <- 100
````

Below is the sequential version with a for loop. The `system.time()` function is used to measure how long it takes to process the workload.

````r
# This code segment shows us how long it takes to run on one core.
system.time(
for (i in 1:iters) {
  to.ls <- rnorm(1e6)
  to.ls <- summary(to.ls)
}
)
````

This is the parallel example with the `doParallel` package.

````r
# This code segment shows us how long it takes to run on all available cores.
library(doParallel)

# Obtain the number of cores available.
ncpu <- detectCores()
ncpu

system.time({
  cl <- makeCluster(ncpu)
  registerDoParallel(cl)
  ls<-foreach(icount(iters)) %dopar% {
    to.ls<-rnorm(1e6)
    to.ls<-summary(to.ls)
  }
  stopCluster(cl)
})
````
