## **1) 1-D arrays (Vectors)**

```python
#@markdown ### **a) np.arange() : Increment**
#@markdown >  Let's start with simple ones. Similar to range().
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html

myAry = np.arange(10)
#myAry = np.arange(10, dtype="complex")
#myAry = np.arange(-0.5, 0.5, 0.1)

print("myAry = ", myAry)
```

```python
#@markdown ### **b) np.linspace() : Linear / equal spacing**
#@markdown > Very useful in setting up coordinates, etc.
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

myAry = np.linspace(0, 10, 5)
#myAry = np.linspace(0, 10, 5, endpoint=False)
#myAry = np.linspace(0, 10, 5, dtype="int")
#myAry, step = np.linspace(0, 10, 5, retstep=True)

print("myAry = ", myAry)
```

```python
#@markdown ### **c) np.logspace() : Log spacing**
#@markdown > Still equal spacing, similar usage, but in log scale
# https://numpy.org/doc/stable/reference/generated/numpy.logspace.html

myAry = np.logspace(0, 10, 5)
#myAry = np.logspace(0, 10, 5, base=2)

print("myAry = ", myAry)
```

```python
#@markdown > **Plot it: linear vs. log**
#@markdown > - (Just showing. Don't worry if you do not know matplotlib yet.)

import matplotlib.pyplot as plt
plt.plot(np.logspace(0, 1, 5),'r-s')
plt.plot(np.linspace(0, 1, 5),'b-s')
plt.show()
```

## **2) 2-D arrays (Matrices)**

```python
#@markdown ### **a) np.identity(): Identity matrix**
# https://numpy.org/doc/stable/reference/generated/numpy.identity.html

myAry = np.identity(5)
#myAry = np.identity(5, dtype="int")

print("myAry = \n", myAry)
```

```python
#@markdown ### **a) np.eye(): Identity matrix, but more versatile**
# https://numpy.org/doc/stable/reference/generated/numpy.eye.html

myAry = np.eye(5)
#myAry = np.eye(5, dtype="int")
#myAry = np.eye(5, 7)
#myAry = np.eye(5, k=1)

print("myAry = \n", myAry)
```

```python
#@markdown ### **c) np.diag(): Diagonal matrix**
# https://numpy.org/doc/stable/reference/generated/numpy.diag.html

myAry = np.diag([1, 4, 9, 16, 25])
#myAry = np.diag([1, 4, 9, 16, 25], k=-1)

print("myAry = \n", myAry)
```
