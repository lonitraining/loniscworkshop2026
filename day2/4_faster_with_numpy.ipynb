#### **Question**: How do you add 5 to each element of an 100,000-element integer array?

````python
#@markdown * How to do that with lists?

myList1 = list(range(100000))
myList2 = [val+5 for val in myList1]
print("Result: \n", myList2)
````

````python
#@markdown * How to do that with NumPy arrays?

myAry1 = np.arange(100000)
myAry2 = myAry1 + 5
print("Result: \n", myAry2)
````

````python
#@markdown - Which one is faster?

# Use the magic word: %timeit  (works for IPython & Jupyter notebooks)
# Don't worry if you do not know it

myList1 = list(range(100000))
tList = %timeit -o myList2 = [val+5 for val in myList1]

myAry1 = np.arange(100000)
tAry = %timeit -o myAry2 = myAry1 + 5

speedup = tList.average / tAry.average
print("Speedup = %.1f times" % (speedup))
````

## **1) Universal Functions (ufunc)**

### **a) What is ufunc?**

> - Functions accept **scalar** input also accept **NumPy arrays** exactly the same way.
> - So that you do not need loop!

````python
#@markdown **Example: Calculating square root using np.sqrt().**

# When input is a scalar (number)
x = 25
print("    x      = ", x)
print("np.srqt(x) = ", np.sqrt(x))

# When input is an array
x = np.array([1, 4, 9, 16, 25])
print("\n    x      = ", x)
print("np.srqt(x) = ", np.sqrt(x))
````

> **Available ufuncs:**
> * Arithmetic operators
>   * `+`, `-`, `*`, `/ `
>   * `**` (power), `//` (integer division), `%` (modulus)
> * Bitwise operators
>   * `&` (AND), `|` (OR), `^` (XOR), `~` (NOT)
>   * `<<` (left shift), `>>` (right shift)
> * Assignment operators
>   * w/ arithmetic: `+=`, `-=`, `*=`, `/=`, `**=`, `//=`, `%=`
>   * w/ bitwise: `&=`, `^=`, `|=`, `~=`, `<<=`, `>>=`
> * Comparison operators
>   * `>`, `<`, `>=`, `<=`, `==`, `!=`, `<>`
> * Mathematical functions
>   * `np.sin()`, `np.cos()`, `np.tan()` ...
>   * `np.exp()`, `np.log()`, `np.log10()` ...
>   * `np.abs()`, `np.sqrt()` ...
> * Special Functions
>   * `scipy.special.*`
> * . . . and many, many more.

### **b) Examples**

## **3) Aggregation Functions**

### **a) What is aggregation function?**

* Aggregations are functions which summarize the values in an array (e.g. min, max, sum, mean, etc.)
* Numpy aggregations are much faster than Python built-in functions
* Available functions:
  * Sort, search and counting
    * `min()`, `max()`, `argmin()`, `argmax()`, ...
    * `sum()`, `prod()` (calculate product instead of sum), ...
    * `sort()`, ...
  * Statistics
    * `mean()` (arithmetic average), `average()` (weighted average), `median()` (median value), ...
    * `var()` (variance), `std()` (standard deviation, square root of variance), ...
  * . . . and many, many more.
* The specific syntax are different for different functions. Here we just give you a glimpse how they work. For details, see https://numpy.org/doc/stable/reference/routines.html

### **b) Examples**

````python
#@markdown **Basic usage (E.g., "sum()")**

# Create a 3x3 matrix
myAry = np.arange(9).reshape(3,3)
print("myAry = \n", myAry)

# Two ways to use (works for most of them):
print("\nnp.sum(myAry) = ", np.sum(myAry))  # As a function
print("\nmyAry.sum() = ", myAry.sum())      # As a method
````

````python
#@markdown **Sometimes, function and method forms work slightly differently (E.g., "sort()")**

# Create an array of random values
myAry = np.random.random(10)
print("myAry = \n", myAry)

# As function: Returns a *copy* of the result
print("\nnp.sort(myAry) = \n", np.sort(myAry))

# As method: *in-place*
myAry.sort()
print("\nmyAry.sort() = \n", myAry)
````

````python
#@markdown **Limiting on specifying axes**

# Create a 3x3 matrix
myAry = np.arange(9).reshape(3,3)
print("myAry = \n", myAry)

print("\nmyAry.sum() = \n", myAry.sum())              # Sum all
print("\nmyAry.sum(axis=0) = \n", myAry.sum(axis=0))  # Sum axis 0 (rows), keep axis 1 (columns)
print("\nmyAry.sum(axis=1) = \n", myAry.sum(axis=1))  # Sum axis 1 (columns), keep axis 0 (rows)
````

````python
#@markdown **Return elements or indeces (E.g., "max()" and "argmax()")**

# Create an array of random values
myAry = np.random.random(10)
print("myAry = \n", myAry)

print("\nmyAry.max() = \t", myAry.max())        # Return the maximum value
print("\nmyAry.argmax() = \t", myAry.argmax())  # Returns the argument (index) that contains the maximum value
````

````python
#@markdown **Let's try some statistics**

myAry = np.arange(10).reshape(2,5)
print("myAry = \n", myAry)

# Arithmetic average
print("\nArithmetic average: \n", myAry.mean())
print("\nArithmetic average over the column: \n", myAry.mean(axis=1))

# Standard deviation
print("\nStandard deviation: \n", myAry.std())
print("\nStandard deviation over the column: \n", myAry.std(axis=1))
````

## **4) Advanced Indexing**

### **a) What is advanced indexing?**:

- **Basic indexing**
  - Passing: **ONE** integer as an index
  - Return:  **ONE** element
  - "***Scalar in, scalar out*** "
>```
>myAry = np.arange(10)
>print(myAry[1])
>```

- **Advanced indexing**
  - Passing: an **array** as an index.
  - Return: **Same shape** of the array index
  - "***Array in, array out*** "
>```
>myAry = np.arange(10)
>indAry = np.array([1, 3, -2])
>print(myAry[indAry])
>```

- For details: https://numpy.org/doc/stable/user/basics.indexing.html#advanced-indexing

### **b) Integer Array Indexing**

- Passing **arrays of integers** as indeces.

````python
#@markdown **Use integer array in the place of integer**

myAry = np.arange(10, 0, -1)
print("myAry = ", myAry)

# i)   Integer as index
print("\nScalar in, scalar out:\n", myAry[3])

# ii)  1-D array as index
indAry = np.array([1, 3, -2])
print("\n1-D array in, 1-D array out:\n", myAry[indAry])

# iii) 2-D array as index
indAry = np.array(
    [[4,3],
     [2,-1]])
print("\n2-D array in, 2-D array out:\n", myAry[indAry])
````

>  * **TIP**:
>    * You should always use **NumPy arrays** in integer array indexing.
>    * I advice against using **lists**. The behavior is inconsistent because of some deprecated syntax.

````python
#@markdown **You can use integer array indexing on multi-dimensional array**

myAry = np.arange(25, 0, -1).reshape(5,5)
print("myAry = \n", myAry)

# Integers as indeces
print("\nScaler in, scalar out:\n", myAry[3,4])

# 1-D arrays as indeces
xAry_1 = np.array([1, 3, -2])
yAry_1 = np.array([0, -1, 1])
print("\n1-D array in, 1-D array out:\n", myAry[xAry_1, yAry_1])

# 2-D arrays as indeces
xAry_2 = np.array(
    [[4,3],
     [2,-1]])
yAry_2 = np.array(
    [[-3,1],
     [-2,0]])
print("\n2-D array in, 2-D array out:\n", myAry[xAry_2, yAry_2])
````

````python
#@markdown **But! The input shapes of different axis have to be COMPATIBLE**

myAry = np.arange(25, 0, -1).reshape(5,5)
print("myAry = \n", myAry)

# Using two incompatible integer arrays
xAry = np.array([1, 3, -2])
yAry = np.array(
    [[-3,1],
     [-2,0]])

print(myAry[xAry, yAry])

#@markdown *Think again, what does "compatible" here mean?*
````

````python
#@markdown **But! Is this flattened result a copy or a reference?**
#@markdown > * *Spoiler alert! You may need it for this afternoon's exercises*

# First, as mentioned above, n-D boolean array indexing will attempt to flatten
myAry = np.arange(25, 0, -1).reshape(5,5)
mask = myAry > 12
print(" myAry = \n", myAry)
print("\nmyAry > 12 = \n", a>b)
print("\nmyAry[myAry>12] = \n", myAry[mask])

# Now, is "myAry[mask]" a copy or a reference?
myAry[mask] = 0
print("\nNow: \nmyAry = \n", myAry)
````

> *Indexing and slicing are **ALL REFERENCES**! No matter how unintuitive it seems*

````python
#@markdown **A useful example: Picking out bad data points**

# Find the mean value of all numbers in this array:
myAry = np.array([[1., 2.],
                  [np.nan, 3.],
                  [4., np.nan]])
print("myAry = \n", myAry)

# Not what I want
print("\n<Not what I want>")
print("Mean: ", myAry.mean())

# Is what I want
print("\n<Is what I want>")
mask = ~np.isnan(myAry)
print("\nIs it a number?:\n", mask)
print("\nmyAry[mask] = ", myAry[mask])
print("Mean: ", myAry[mask].mean())
````
